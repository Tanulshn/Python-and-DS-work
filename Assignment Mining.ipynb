```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage
wardlink = linkage(bank_scaled, method = 'average' )
```

```python
plt.figure(figsize=(20,10))
dend =dendrogram(wardlink)
```

```python
# Lets simplify the dendrogram
dend = dendrogram(wardlink, 
                  truncate_mode = 'lastp',
                 p = 10)
```

```python
# Converting the dendrogram to clusters
# Import module
from sklearn.cluster import AgglomerativeClustering
```

```python
# Set n_clusters, affinity (method of distance measurement (euclidean, manhattan, ....))

agg = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
cluster_agg = agg.fit_predict(bank_scaled)
cluster_agg
```

```python
bank['Agglo-Clusters']= cluster_agg
```

```python
# Create Table
bank_agg = bank.groupby('Agglo-Clusters').mean().round(2)

bank_agg.drop(['KM Cluster'], axis =1, inplace = True)
bank_agg
```
